```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y, marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Display the plot
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure and axes
fig, ax = plt.subplots()

# Create a line plot using ax.plot()
ax.plot(x, y, marker='o')

# Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line Plot using ax.plot()')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

num_samples = 100
age = np.random.randint(18, 65, size=num_samples)
medical_cost = 1000 + 15 * age + np.random.normal(0, 200, size=num_samples)

# Create a DataFrame from the dummy data
df = pd.DataFrame({'age': age, 'medical_cost': medical_cost})

# Define the range of ages for the linear regression line
age_range = np.linspace(df['age'].min(), df['age'].max(), num=100)

# Define a simple linear regression function
def lin_reg(x):
    return 1000 + 15 * x

# Create the scatter plot with the linear regression line
ax = df.plot.scatter(x='age', y='medical_cost', color='b', label='Data Points')
ax.plot(age_range, lin_reg(age_range), color='r', label='Linear Regression Line')
ax.set_title('Medical Costs vs. Age')
ax.set_xlabel('Age')
ax.set_ylabel('Medical Cost')
```

```python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2)

# Now, you can plot on each subplot using 'axes'
axes[0, 0].plot([1, 2, 3, 4], [1, 4, 9, 16])
axes[0, 0].set_title('Subplot 1')

axes[0, 1].scatter([1, 2, 3, 4], [1, 4, 9, 16])
axes[0, 1].set_title('Subplot 2')

axes[1, 0].bar(['A', 'B', 'C', 'D'], [10, 20, 15, 30])
axes[1, 0].set_title('Subplot 3')

axes[1, 1].hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
axes[1, 1].set_title('Subplot 4')

# Adjust spacing between subplots
plt.tight_layout()
```

#### Bar charts

```python
# Sample data
categories = ['Category A', 
								'Category B', 
								'Category C', 
								'Category D']
counts = [10, 25, 15, 30]

# Create a basic bar chart
plt.bar(categories, counts)

# Adding labels
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Categorical Data Bar Chart')
```

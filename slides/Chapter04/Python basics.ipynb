```python
# Set n and total equal to zero
n = 20
total = 0

# Add the numbers 1 up to n to total
for i in range(1, n + 1):
    total += i
    
# Print the result and n * (n+1) / 2
print(total, ' - ', n * (n + 1) / 2)
```

```python
# We check if n * (n + 1) / 2 is equal to the sum of 1 up to n.

n = 20 # We validate this for n = 20 here
total = 0

for i in range(1, n + 1):
    total += i
    
print(total, ' - ', n * (n + 1) / 2)
```
